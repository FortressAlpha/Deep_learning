```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import LabelEncoder , StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import Perceptron

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import Flatten

from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Dropout

from tensorflow.keras.utils import to_categorical
```

```python
df = pd.read_csv('mnist_train.csv')
df_test = pd.read_csv('mnist_test.csv')
```

```python
df.head()
```

```python
x_train = df.drop("label",axis=1).values
y_train = df["label"].values
x_test = df_test.drop("label",axis=1).values
y_test = df_test["label"].values
```

```python
x_train = x_train.astype('float32')/255.0
x_test = x_test.astype('float32')/255.0
```

```python
x_train_img = x_train.reshape(-1,28,28)
x_test_img = x_test.reshape(-1,28,28)
```

```python
x_test_img
```

```python
y_train_cat = to_categorical(y_train,10)
y_test_cat = to_categorical(y_test,10)
```

```python
perceptron = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(10,activation='softmax')
])
```

```python
perceptron.compile(optimizer="sgd",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
history_percp = perceptron.fit(x_train_img,y_train_cat,epochs=10,batch_size=32,validation_data=(x_test_img,y_test_cat),verbose=1)
```

```python
acc_percp = perceptron.evaluate(x_test_img,y_test_cat,verbose =0)[1]
```

```python
acc_percp
```

```python
# ANN
```

```python
ann = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128,activation="relu"),
    Dense(64,activation="relu"),
    Dense(10,activation="softmax")
])
```

```python
ann.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
history_ann = ann.fit(x_train_img,y_train_cat,epochs=10,batch_size=32,validation_data=(x_test_img,y_test_cat),verbose=1)
```

```python
acc_ann = ann.evaluate(x_test_img,y_test_cat,verbose =0)[1]
```

```python
acc_ann
```

```python
#CNN
```

```python
x_train_cnn = x_train.reshape(-1,28,28,1)
x_test_cnn = x_test.reshape(-1,28,28,1)
```

```python
cnn = Sequential([
    Conv2D(32,kernel_size=(3,3),activation="relu",input_shape=(28,28,1)),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(64,kernel_size=(3,3),activation="relu"),
    MaxPooling2D(pool_size=(2,2)),
    Flatten(),
    Dense(128,activation="relu"),
    Dropout(0.5),
    Dense(10,activation="softmax")
])
```

```python
cnn.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
history_cnn = cnn.fit(x_train_img,y_train_cat,epochs=10,batch_size=32,validation_data=(x_test_img,y_test_cat),verbose=1)
```

```python
acc_cnn = cnn.evaluate(x_test_img,y_test_cat,verbose =0)[1]
```

```python
acc_cnn
```

```python

def show_side_by_side(models, model_names, X, X_cnn, y_true, n=5):
    idxs = np.random.choice(len(X), n, replace=False)
    plt.figure(figsize=(15, 6))
    for i, idx in enumerate(idxs):
        plt.subplot(2, n, i+1)
        plt.imshow(X[idx].reshape(28, 28), cmap="gray")
        plt.axis("off")
        plt.title(f"True: {y_true[idx]}")
        preds = [np.argmax(model.predict(X_cnn[idx].reshape(1, 28, 28, 1) if name == "CNN" else X[idx].reshape(1, 28, 28)))
                 for model, name in zip(models, model_names)]
        plt.subplot(2, n, n+i+1)
        plt.axis("off")
        plt.title("\n".join(f"{n}: {p}" for n, p in zip(model_names, preds)))
    plt.tight_layout()
    plt.show()
```

```python
show_side_by_side([perceptron, ann, cnn], ["Perceptron", "ANN", "CNN"], x_test_img, x_test_cnn, y_test, 5)
```
